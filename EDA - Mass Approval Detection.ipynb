```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
packt_palette = ["#f97141","#535555", "#89452D", "#1a1a1a",  "#573428",    "#E4E7ED"]

def custom_palette(custom_colors):
    packt = sns.set_palette(sns.color_palette(custom_colors))
    sns.palplot(sns.color_palette(custom_colors),size=0.8)
    plt.tick_params(axis='both', labelsize=0, length = 0)

custom_palette(packt_palette)
```

```python
df = pd.read_csv('datasets/combined_data.csv')
```

```python
print('The shape of dataset is',df.shape)
print('\nThe features in dataset are:\n----------------------------------------------------------------------------\n',df.columns)
```

## Bar Plot for most common time intervals for request approvals

```python
df['approved_on'] = pd.to_datetime(df['approved_on'])
df['requested_on'] = pd.to_datetime(df['requested_on'])

df['time_taken'] = (df['approved_on']  - df['requested_on']).dt.total_seconds()/(3600)
```

```python
df['time_taken'] = df['time_taken'].round(decimals = 0)
```

```python
df['time_taken'].value_counts()
```

```python
plt.figure(figsize=(10,7))
df['time_taken'].value_counts()[:30].plot(kind = 'bar', color=packt_palette)
plt.title('Bar Plot of 30 most common time intervals for request approvals', size=15)
plt.ylabel('Density', size=13)
plt.xlabel('Time Taken', size=13)
plt.show()
```

- Most of the requests have been attended within **15 hours** from when they were requested.

## Density Plot of time taken

```python
plt.figure(figsize=(10,7))
sns.kdeplot(x = df['time_taken'], color='orange', fill=True, shade=True)
plt.title('Density Plot for Time Taken for request approval', size=15)
plt.ylabel('Density', size=13)
plt.xlabel('Time Taken', size=13)
plt.show()
```
